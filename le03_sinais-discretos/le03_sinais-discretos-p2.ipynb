```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal # <- Biblioteca de Processamento de sinais
import sys
sys.path.append('../lib/') # <- Diretório de funções

from myAudioProcessingLib import *
```

```python
# Definição de diretórios
SOUNDS_PATH = '../sounds/' # Pasta de arquivos de som
SOUND_FILE = 'freddy.wav' # Sinal de entrada

# Pasta de arquivos de IR (impulse response)
# Obtidas de: https://www.voxengo.com/impulses/
IR_PATH = '../impulse-response/'

IR_FILE = 'Direct Cabinet N1.wav' # Resposta ao impulso
# IR_FILE = 'Cement Blocks 1.wav' # Resposta ao impulso
# IR_FILE = 'Chateau de Logne, Outside.wav' # Resposta ao impulso
```

```python
# 1. PASSO 1: Carregamos o nosso sinal de entrada
# x : Sinal x[n]
# fs: Frequencia de amostragem
x, fsx = audioread(SOUNDS_PATH + SOUND_FILE)
```

```python
# Conversao de audio stereo (2D) para monoaural (1D)
# sinal 2D (canal L: esquerdo, canal R: direito) -> sinal 1D (canal unico)
if x.ndim > 1:
    x = x.sum(axis=1) / 2.0
```

```python
# Criamos o vetor de amostras 'n' de x[n]
nx = np.arange(len(x))

# Tempo de amostragem (vetor de tempo x[n])
tx = nx / fsx
```

```python
print('x[n] tem '+str(nx[-1])+' amostras, fs='+str(fsx)+' Hz (Hertz) e duracao de '+str(np.round(tx[-1],1))+' segundos')
```

```python
# 2. PASSO 2: Carregamos a "resposta ao impulso"
h, fsh = audioread(IR_PATH + IR_FILE)

# Conversao stereo -> mono
if h.ndim > 1:
    h = h.sum(axis=1) /2.0

# Criamos o vetor de amostras 'n' de h[n]
nh = np.arange(len(h))
th = nh / fsh # vetor de tempo h[n]
```

```python
# 3. PASSO 3: Convolucao de sinais
y = signal.convolve(x, h)
ny = np.arange(len(y)) # vetor de amostras 'n' de y[n]
ty = ny / fsx # vetor de tempo y[n]
```

```python
# Plots
plt.figure(), plt.plot(tx, x, 'b'), plt.xlabel(r'$n$'), plt.ylabel(r'$x[n]$'), plt.title('Sinal de entrada'), plt.plot()
plt.figure(), plt.plot(th, h, 'm'), plt.xlabel(r'$n$'), plt.ylabel(r'$h[n]$'), plt.title('Resposta ao impulso'), plt.plot()
plt.figure(), plt.plot(ty, y, 'r'), plt.xlabel(r'$n$'), plt.ylabel(r'$y[n]$'), plt.title('Sinal de saida'), plt.plot()
```
