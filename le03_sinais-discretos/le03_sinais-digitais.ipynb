# Sinais digitais Essenciais

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Sinal delta
n     = np.arange(N)
xn    = np.zeros(N)
xn[7] = 1 # x[n-7]

# Versao 2
xn    = [0] * N
xn[0] = 1
plt.stem(xn)
```

```python
# 2. Degrau
n      = np.arange(N)
un     = np.zeros(N)
un[8:] = 1

# Segunda versao
un     = [0] * int(N/2) + [8] * int(N/2)
plt.stem(n, xn)
```

```python
# 3. Exponencial
a  = 0.5
yn = np.abs(a)**n * un

plt.stem(n, yn)
```

```python
# 4. Sinal periodico
N  = 100 # numero de amostras
n  = np.arange(N) - 7 # vetor de amostras

A   = 10
P   = 53 # periodo do sinal
wo  = 2*np.pi / P
phi = np.pi/2 # fase
xn  = A * np.sin(wo * n + 0)
yn  = A * np.cos(wo * n + 3*np.pi/2)

# Plotagem
plt.stem(n, xn, 'ro')
plt.stem(n, yn, 'bd')
# plt.xlim([0, 20])
```

```python
# 5. Sinal aleatorio
# Distribuicoes: Normal (Gaussian), Poisson, binomial, uniforme, ...
N = 10000
n  = np.arange(N) - 7 # vetor de amostras

# xn = np.random.rand(N) # uniforme
xn = np.random.randn(N) # Gaussiano (normal)

# Plot
fig, ax = plt.subplots(1)
ax.stem(n, xn, 'ro')
plt.xlim([0, 100])
plt.title('Sinal x[n] aleatorio')

fig, ax = plt.subplots(1)
ax.hist(xn, int(N/100))
plt.title('Distribuicao de x[n] aleatorio')
```
